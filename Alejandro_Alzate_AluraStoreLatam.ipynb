### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

```python
tienda1['Tienda'] = 'Tienda_1'
tienda2['Tienda'] = 'Tienda_2'
tienda3['Tienda'] = 'Tienda_3'
tienda4['Tienda'] = 'Tienda_4'
tienda1.head()
```

```python
df = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
df
```

#1. Análisis de facturación



```python
ventas_por_tienda = df.groupby('Tienda')['Precio'].sum()
ventas_por_tienda.map('{:,.1f}'.format)
```

```python
total_ventas = sum(ventas_por_tienda)
print(f'El total de ventas fue: ${total_ventas}')
```

# 2. Ventas por categoría

```python
ventas_categoria = df.groupby(['Tienda', 'Categoría del Producto'])['Precio'].agg(['sum', 'count'])
ventas_categoria.rename(columns={'sum': 'Venta Categoria', 'count': 'Cantidad'}, inplace=True)
ventas_categoria
```

```python
ventas_categoria_general = ventas_categoria.groupby('Categoría del Producto')[['Cantidad','Venta Categoria']].sum()
ventas_categoria_general['Venta Categoria'] = ventas_categoria_general['Venta Categoria'].map('{:.1f}'.format)
ventas_categoria_general
```

```python
cantida_total_categoria = sum(ventas_categoria['Cantidad'])
valor_total_categoria = sum(ventas_categoria['Venta Categoria'])

print(f'La cantidad total de ventas en todas las tiendas fue de: {cantida_total_categoria}, por un valor de ${valor_total_categoria}')
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = round(df.groupby('Tienda')['Calificación'].mean(),2)
calificacion_promedio
```

# 4. Productos más y menos vendidos

Productos más vendidos


```python
# Calculamos la cantidad de cada producto
cantidad_productos = df.groupby('Tienda')['Producto'].value_counts().reset_index(name='count')

# La cantidad máxima por cada tienda
max_cantidad = cantidad_productos.groupby('Tienda')['count'].max().reset_index()

# Los unimos en un solo DataFrame filtrando los productos más vendidos
productos_mas_vendidos = pd.merge(cantidad_productos, max_cantidad, on=['Tienda', 'count'])

productos_mas_vendidos.rename(columns={'count': 'Cantidad'}, inplace=True)
productos_mas_vendidos[['Tienda', 'Producto', 'Cantidad']]
```

Productos menos vendidos

```python
#Obtenemos la cantidad mínima por cada tienda
min_cantidad = cantidad_productos.groupby('Tienda')['count'].min().reset_index()

# Los unimos en un solo DataFrame filtrando los productos menos vendidos
productos_menos_vendidos = pd.merge(cantidad_productos, min_cantidad, on=['Tienda', 'count'])


productos_menos_vendidos.rename(columns={'count': 'Cantidad'}, inplace=True)
productos_menos_vendidos[['Tienda', 'Producto', 'Cantidad']]
```

Productos con mayor ganancia

```python
#También podemos crear un top-3 de los productos con mayor ganancia, agrupados por tienda

productos_tienda = df.groupby(['Tienda', 'Producto'])['Precio'].agg(['sum', 'count'])
productos_tienda = productos_tienda.sort_values(by='sum', ascending=False)
top3_productos_tienda = productos_tienda.groupby('Tienda', group_keys=True).head(3).sort_values(by='Tienda')
top3_productos_tienda.rename(columns={'sum': 'Ganancia', 'count': 'Cantidad'}, inplace=True)
top3_productos_tienda
```

Productos con menor ganancia

```python
#También podemos crear un grupo de los 3 productos con menor ganancia, agrupados por tienda

productos_tienda_menos = productos_tienda.sort_values(by='sum')
menos3_productos_tienda = productos_tienda_menos.groupby('Tienda', group_keys=True).head(3).sort_values(by='Tienda')
menos3_productos_tienda.rename(columns={'sum': 'Ganancia', 'count': 'Cantidad'}, inplace=True)
menos3_productos_tienda
```

```python
import matplotlib.pyplot as plt
import pandas as pd



# Convertir la columna 'Fecha de Compra' a formato fecha
df2['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'])

# Crear figura y subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))  # Adjust figsize as needed
fig.suptitle('Timeseries of Precio by Año de Compra for Each Tienda', fontsize=16)
colores = ['blue', 'green', 'red', 'purple']

# Iterar cada Tienda y crear las gráficas de series temporales
for i, tienda in enumerate(['Tienda_1', 'Tienda_2', 'Tienda_3', 'Tienda_4']):
    row = i // 2  # Calculate row index for subplot
    col = i % 2   # Calculate column index for subplot

    # Filtrar por Tienda
    tienda_data = df2[df2['Tienda'] == tienda]


    timeseries_data = tienda_data.groupby(tienda_data['Fecha de Compra'].dt.year)['Precio'].sum()

    # Crear la gráfica
    axs[row, col].plot(timeseries_data.index, timeseries_data.values, color=colores[i])
    axs[row, col].set_title(tienda)
    axs[row, col].set_xlabel('Año de Compra')
    axs[row, col].set_ylabel('Precio')

    axs[row, col].set_xticks([2020, 2021, 2022, 2023])

# Ajustar el espacio y mostrar la gráfica
plt.tight_layout()
plt.show()
```
